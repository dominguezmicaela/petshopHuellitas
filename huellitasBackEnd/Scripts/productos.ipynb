**Imputacion de valores nulos en el campo precio**



Se decide imputar los valores nulos por la mediana ya que conserva mejor la tendencia central de los datos y evita sesgos adicionales en comparacion de la media que esta inflada por los valores atipicos.



```python
mediana= df_productos['precio'].median()
df_productos['precio']=df_productos['precio'].fillna(mediana)
```

```python
#verificacion
df_productos['precio'].isnull().sum()
```

```python
import random

def asignarPrecioLogico(fila):
    # Convertimos el nombre a min√∫scula para buscar f√°cil
    nombre = str(fila['nombre']).lower()

    if any(x in nombre for x in ['cama', 'transportadora', 'casa', 'jaula']):
        return random.randint(40000, 95000)
    elif any(x in nombre for x in ['comedero', 'alimento', 'shampoo', 'cuidado']):
        return random.randint(15000, 35000)
    elif any(x in nombre for x in ['juguete', 'snack', 'correa', 'collar', 'accesorio']):
        return random.randint(3500, 12500)
    else:
        return random.randint(5000, 20000)

df_productos['precio'] = df_productos.apply(asignarPrecioLogico, axis=1)

display(df_productos[['nombre', 'precio']].head(10))
```

# StockActual

El stock actual va a usar datos del campo availability pero como tiene datos sucios uso el siguiente criterio:


*  si esta en stock el producto se va a llenar con un numero aleatoorio entre 3 y 50
*   si no hay datos o agotado 0



```python
import random
```

```python
# Defino una funcion para generar stock realista
def calculoStock(texto):
  if pd.isna(texto):
    return 0
  if "stock" in texto.lower():
    return random.randint(3,50)
  else:
    return 0
```

```python
#Creo la columa stockActual y uso la funcion definida previamente
df_productos['stockActual']=df_productos['availability'].apply(calculoStock)
```

```python
def generadorNombre(texto):
    if pd.isna(texto): return "Accesorio General"
    texto = str(texto).lower()

    # 1. QUE PRODUCTO ES (Detectamos lo importante)
    producto = "Accesorio"
    if any(x in texto for x in ['food', 'nutrition', 'diet']): producto = "Alimento"
    elif any(x in texto for x in ['snack', 'treat', 'biscuit', 'cookie']): producto = "Snacks"
    elif any(x in texto for x in ['toy', 'ball', 'chew', 'squeak', 'wand']): producto = "Juguete"
    elif any(x in texto for x in ['bed', 'mat', 'blanket', 'hammock']): producto = "Cama"
    elif any(x in texto for x in ['collar', 'leash', 'harness']): producto = "Correa/Collar"
    elif any(x in texto for x in ['shampoo', 'flea', 'tick']): producto = "Cuidado"
    elif any(x in texto for x in ['bowl', 'feeder']): producto = "Comedero"
    elif any(x in texto for x in ['house', 'cage', 'carrier']): producto = "Transportadora"
    # AC√Å SALVAMOS A LA PALA (Higiene)
    elif any(x in texto for x in ['scoop', 'litter', 'poop', 'waste', 'bandeja', 'pala']): producto = "Higiene"

    # 2. PARA QUE ANIMAL ES
    animal = ""
    if any(x in texto for x in ['cat', 'kitten']): animal = "para Gatos"
    elif any(x in texto for x in ['dog', 'puppy']): animal = "para Perros"
    elif 'bird' in texto: animal = "para Aves"

    # 3. ARMAMOS EL NOMBRE (Simple y al pie)
    # Quedar√° algo como: "Juguete para Gatos" o "Higiene para Gatos"
    nombre_final = f"{producto} {animal}"

    return " ".join(nombre_final.split())

# Aplicamos la funci√≥n al DataFrame
df_productos['nombre'] = df_productos['title'].apply(generadorNombre)
display(df_productos[['nombre']].head(10))
```

# idCategoria

```python
#clasificacion del 1 al 5
def clasificarCategoria(texto):
    if pd.isna(texto):
        return 1

    texto = str(texto).lower()
    #farmacia
    palabras_farmacia = ['medic', 'health', 'vitamin', 'flea', 'tick', 'care', 'supplement', 'shampoo', 'dental']
    if any(p in texto for p in palabras_farmacia):
        return 3
    #snacks
    palabras_snack = ['treat', 'snack', 'biscuit', 'cookie', 'chew', 'bone', 'jerky']
    if any(p in texto for p in palabras_snack):
        return 5
    #accesorios
    palabras_accesorios = ['bed', 'toy', 'collar', 'leash', 'harness', 'bowl', 'carrier', 'cloth', 'apparel', 'house', 'cage', 'scratch', 'post']
    if any(p in texto for p in palabras_accesorios):
        return 4
    #gatos
    if any(p in texto for p in ['cat', 'kitten', 'feline']):
        return 2
    #aves
    palabras_aves = ['bird', 'parrot', 'canary', 'budgie', 'cockatiel', 'avian', 'parakeet', 'finch', 'macaw']
    if any(p in texto for p in palabras_aves):
        return 6

   #perros
    return 1
```

```python
# le aplico a la columna breadcrumbs la funcion creada previamente para que clasifique los productos
df_productos['idCategoria'] = df_productos['breadcrumbs'].apply(clasificarCategoria)

#verificacion
print("\nConteo de productos por categ:")
print(df_productos['idCategoria'].value_counts().sort_index())

# Vemos una muestra para confirmar
display(df_productos[['breadcrumbs', 'idCategoria']].head(10))
```

# Limpieza y traduccion

```python
columnas_finales = [
    'nombre',
    'descripcion',
    'precio',
    'stockActual',
    'stockMinimo',
    'img',
    'idCategoria'
]
productos= df_productos[columnas_finales].copy()
```

```python
traductor = GoogleTranslator(source='auto', target='es')
def traducir_seguro(texto):
    try:
        if pd.isna(texto) or texto == "":
            return "Sin descripci√≥n"
        # Traducimos
        return traductor.translate(str(texto))
    except Exception as e:
        # Si falla (por internet o lo que sea), devolvemos el original en ingl√©s para no perderlo
        return str(texto)
```

```python
productos['descripcion'] = productos['descripcion'].apply(traducir_seguro)
```

```python
#fase load
#funcion para escapar de las comillas
def escapar_comillas(valor):
    if pd.isna(valor): return "NULL"
    return str(valor).replace("'", "''")

nombre_archivo = 'insertar_productos_postgres_final.sql'

print(f"Generando {nombre_archivo} adaptado a tabla 'producto'...")

with open(nombre_archivo, 'w', encoding='utf-8') as f:

    f.write("-- SCRIPT PET SHOP HUELLITAS --\n")

    # tabla producto
    f.write("-- Carga de Productos --\n")

    for i, row in productos.iterrows():
        nombre = escapar_comillas(row['nombre'])
        desc   = escapar_comillas(row['descripcion'])
        img    = escapar_comillas(row['img'])
        precio = row['precio']
        stock  = row['stockActual']
        stock_min = row['stockMinimo']
        cat    = row['idCategoria']

        sql = f'INSERT INTO producto (nombre, descripcion, precio, img, "stockActual", "stockMinimo", "idCategoria") VALUES (\'{nombre}\', \'{desc}\', {precio}, \'{img}\', {stock}, {stock_min}, {cat});\n'

        f.write(sql)


# descarga
try:
    from google.colab import files
    files.download(nombre_archivo)
except ImportError:
    pass
```
