```python
import pandas as pd
import random
from faker import Faker
from sqlalchemy import create_engine

# Configura Faker para Argentina
fake = Faker('es_AR')

# Reemplaza con tu cadena de conexi贸n de Neon
DATABASE_URL = "tu_cadena_de_conexion_aqui"
engine = create_engine('URLPRIVADA')
```

# L贸gica de Negocio

```python
# defino segmentos
segmentos = ['CatLover', 'DogLover', 'Aves', 'Ex贸ticos']

def generar_dataset_usuarios(n=15):
    lista_usuarios = []
    for _ in range(n):
        segmento_elegido = random.choice(segmentos)

        lista_usuarios.append({
            "nombre": fake.first_name(),
            "apellido": fake.last_name(),
            "email": fake.email(),
            "passwordHash": "argon2_simulado_5544", # Placeholder para seguridad
            "direccion": fake.address().replace('\n', ', '),
            "segmentoCliente": segmento_elegido,
            "idRol": 2
        })
    return pd.DataFrame(lista_usuarios)

# Generamos los datos en memoria
df_usuarios = generar_dataset_usuarios(500)

# Mostramos un resumen para verificar
print(" Previsualizaci贸n de los usuarios a cargar:")
print(df_usuarios[['nombre', 'apellido','segmentoCliente']].head())
```


# Carga a la base de datos


```python

nombre_tabla = 'Usuario'

try:
    df_usuarios.to_sql(nombre_tabla, engine, if_exists='append', index=False)
except Exception as e:
    print(f" Error al cargar en la base de datos: {e}")
```
